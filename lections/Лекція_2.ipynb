```python
print(np.zeros(5)) # нуль вектор
print(np.ones(5)) # одиничний вектор

v1 = np.array([1, 2, 3])
print(v1 * -1 ) #зворотній
print(np.linalg.norm(v1)) #норма

v2 = np.array([4,5,6])
print(f'Equal?: {np.array_equal(v1, v2)}')
print(f'Collinear?: {np.allclose(np.zeros_like(v1), np.cross(v1, v2))}')

a = np.array([7, 8, 9])
matrix = np.vstack((v1, v2, a))
det = np.linalg.det(matrix)
print(f'Complain?: {np.isclose(det, 0)}')
```

```python
# Додавання/віднімання скаляра та вектора
a = np.array([1, 2, 3, 4, 5], dtype=float)
print(a + 2)  # [3. 4. 5. 6. 7.]
print(a - 1)  # [0. 1. 2. 3. 4.]

# Множення на скаляр
b = 5*a
print(b)# [ 5. 10. 15. 20. 25.]

# Додавання та віднімання двох векторів
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)  # [5 7 9]
print(a - b)  # [-3 -3 -3]
# Скалярний добуток
print(np.dot(a, b))  # 32

# По елементне множення і ділення
print(a * b)  # [ 4 10 18]
print(b / a)  # [4.  2.5 2. ]
```

```python
import numpy as np

a = np.array([[1, 2, 3],
              [0, 1, 2]])

b = np.array([[ 1, 2],
              [ 0, 4],
              [-1, 0]])

print(a.shape, b.shape)
# множення матриць
print('multi:\n', np.dot(a, b))
# Транспонування матриць
print('transp:\n',a.T)
# Зміна розмірності
print('reshape:\n',a.reshape((3, 2)))
# Плоска матриця
print('flatten:\n',a.flatten())
# Одинична матриця
print('eye:\n',np.eye(4, k=1, dtype=float))
#Обернена матриця
a1=np.array([[1, 2],
              [3, 4]])
print('invariant:\n',np.linalg.inv(a1))
```

```python
# Для створення масиву, що містить послідовність чисел, існують два методи: np.arange та np.linspace
#Метод np.arange аналогічний методу Python range
import numpy as np

a = np.arange(5)
b = np.arange(1, 3, 0.5)
print(a)
print(b)

a = np.linspace(1, 5, num=2)
b = np.linspace(1, 3, num=3)
print(a)
print(b)
```
