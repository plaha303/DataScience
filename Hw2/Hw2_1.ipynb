```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
tables = pd.read_html(data_url,
                      thousands='',
                      decimal=',',
                      match = 'Коефіцієнт народжуваності в регіонах України',
                     )
table = pd.DataFrame(tables[0])
table.head()
```

```python
table.shape
```

```python
table.replace("—", np.nan, inplace=True)
table.head()
```

```python
table.dtypes
```

```python
table.replace("—", np.nan)

numeric_columns = table.select_dtypes(include=['object'])
for column in numeric_columns.columns:
    if column != 'Регіон':
        table[column] = pd.to_numeric(table[column], errors='coerce')

table.dtypes
```

```python
missing_percentage = table.isnull().sum() / len(table) * 100

missing_percentage
```

```python
table = table.drop(table.index[-1])

table
```

```python
means = table.mean(axis=0, numeric_only=True)
table = table.fillna(means)

table
```

```python
means2019 = table.mean(axis=0, numeric_only=True)['2019']

table[table['2019'] > means2019]
```

```python
table[table['2014'] == table['2014'].max()]
```

```python
plt.bar(table["Регіон"], table["2019"], color='skyblue')
plt.title('Народжуваність по регіонах у 2019 році')
plt.xlabel('Регіон')
plt.ylabel('Кількість народжених')
plt.xticks(rotation=90, ha='right')
plt.show()
```
